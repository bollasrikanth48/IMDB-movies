```python

import numpy as np #importing numpy library
import pandas as pd #importing pandas library
#reading the data set file that we have selected for project
df=pd.read_csv('sample_data/final data set-movies.csv')

#1111111111111111111111111111111111111111111111111111111111111111111111111111111111111111
def Top_Director():

  import pandas as pd
  import matplotlib.pyplot as plt

  # Group the data by genre and count the number of entries in each group
  D_counts1 = df.groupby('Directors').size().reset_index(name='count')

  # Sort the genre counts by descending order and select the top 10
  top_D = D_counts1.sort_values('count', ascending=False).head(10)
  # Add an empty string to the beginning of the 'genre' array


  # Create a bar chart of the top 10 genres
  plt.bar(top_D['Directors'], top_D['count'], color=['Red', 'Blue', 'Green', 'Yellow', 'Orange', 'Pink', 'Purple', 'Black', 'Gray', 'Brown'])

  # Add axis labels and a title to the plot
  plt.xlabel('directors')
  plt.ylabel('Count')
  plt.title('Top 10 directors by Count')

  # Rotate the x-axis labels for better readability
  plt.xticks(rotation=45)

  # Display the plot
  plt.show()
  # Group the data by director and count the number of movies for each director
  director_counts = df.groupby('Directors')['Title'].count()
  # Sort the data by the number of movies in descending order
  director_counts = director_counts.sort_values(ascending=False)

  # Display the top directors
  print("\nFollowing are the Top 10 directors\n",director_counts.head(10))

######################################################################################################################

def Top_Genres():

  import pandas as pd
  import matplotlib.pyplot as plt

  genre_counts = df.groupby('Genre').size().reset_index(name='count')

  # Sort the genre counts by descending order and select the top 10
  top_genres = genre_counts.sort_values('count', ascending=False).head(10)

  # Create a pie chart of the genre counts
  plt.pie(top_genres['count'], labels=top_genres['Genre'], autopct='%1.1f%%', colors=['Red', 'Blue', 'Green', 'Yellow', 'Orange', 'Pink', 'Purple', 'Black', 'Gray', 'Brown'])

  # Add a title to the plot
  plt.title('Top 10 Genres by Percentage')

  # Display the plot
  plt.show()

  # counting the each type of genre frequency from the dataset
  genre_counts = df['Genre'].value_counts()

# displaying the all types of genres and count values
  print("\nFollowing are the Top 10 Genres\n",genre_counts.head(10))

##############################################################################################
def top_3_movies(director_name):
 # load the IMDb top 100 movies dataset into a pandas DataFrame
 movies_df = pd.read_csv("sample_data/final data set-movies.csv")

 # prompt the user for a director's name

 # filter the movies DataFrame to include only movies directed by the user input director
 movies_df = movies_df[movies_df['Directors'] == director_name]

 # sort the movies by their rating in descending order
 movies_df = movies_df.sort_values(by='Rating', ascending=False)

 # get the top 3 rated movies
 top3_movies = movies_df.head(3)

 # print the top 10 rated movies
 print(f"Top 3 rated movies by {director_name}:")
 for _, movie in top3_movies.iterrows():
    print(f"{movie['Title']} ({movie['Year']}) - Rating: {movie['Rating']}")


####################################################################################################

def Most_popular_word():
  # Split words from titles
  words = df['Title'].str.split()

  # developing the dictionary to count the frequency of words
  Freq_words= {}
  for Words_title in words:
    for word in Words_title:
        if word in Freq_words:
            Freq_words[word] += 1
        else:
            Freq_words[word] = 1

  # Convert the dictionary to a pandas Series and sort the data in descending order
  Freq_words = pd.Series(Freq_words).sort_values(ascending=False)

  # Display the 10 most common words
  print("\n\nFollowing are the most frequent words use in Titles of top movies\n ",Freq_words.head(10))


####################################################################################################


def Highest_ROI():
  """cleaning the data of budget and income row, we are replacing all the unkown values to the mean values of the data"""
  for col in ['Budget', 'Income']:
    df[col].replace('Unknown', '0', inplace=True)
  """ for finding the average values, we are converting all the types of curriencies into dollars"""
  """conerting values from google"""
  conversion_dict = {
        '$': 1.0,
        '€': 1.07,
        '£': 1.21,
        '₩': 0.00078,
        'C': 0.73,
        '₹': 0.012,
        '¥': 0.0075,
        'A': 0.68,
        'S': 0.095,
        'D': 0.14,
        'N': 0.10
        }
  import re
  # in this code we aree striping each value and converting the value into conversion value that is dollars
  # the following code is processed with erespective of values in conversion_dict dictionary
  for col in ['Budget', 'Income']:
   df[col] = df[col].apply(lambda x: int(re.sub('[^0-9]', '', x)) * conversion_dict.get(x[0], 1))
  #replacing zero values with the mean values of the coloumn
# drop rows where Budget and Income are 0
  df1 = df[~(df == 0).any(axis=1)]
  df2 = df1[df1['Budget'] >= 100]
  df3 = df2[df2['Income'] >= 100]
  #this is swection is iused to calaculate the return of investment(ROI)
  # Create a new column for ROI
  df3['ROI'] = (df3['Income'] - df3['Budget']) / df3['Budget']

  df3.drop(index=1557,axis=0,inplace=True)

  Avg_Roi= df3.groupby('Directors')['ROI'].mean()
  Avg_Roi= Avg_Roi.sort_values(ascending=False).head(10)
  print("\nFollowing are the Average Returns on investment by each Director decreasing order\n",Avg_Roi)
#
  Avg_Roi= df3.groupby('Year')['ROI'].mean()
  Avg_Roi= Avg_Roi.sort_values(ascending=False)
  print("\nFollowing are the Average Returns on investment in each year\n",Avg_Roi)

 ####################################################################################################

def Avg_rating():
  # Group the data by director and calculate the mean rating for each director
  director_ratings = df.groupby('Directors')['Rating'].mean()

  # Sort the data by the mean rating in descending order
  director_ratings = director_ratings.sort_values(ascending=False)

  # Display the top 5 directors with the highest average ratings
  print("\n\nFollowing are the average ratings of the movies of Top Directors",director_ratings.head(10))

  ####################################################################################################

def Popular_actor():
  from collections import Counter
  #6.most common actors in top 100 movies from past 20 years

  # Split the actors column into a list of names
  actors = df['Stars'].str.split(',').explode().str.strip()

  # Count the frequency of each actor
  actor_counts = Counter(actors)

  # Display the top 5 most common actors
  return(actor_counts.most_common(100))

#######################################################################################################

def main():
  while True:
    name=input("\n\nplease enter your name (or press enter to exit): ")
    if name=="":
      print("\n\nBye Byeee see u later...\n  ")
      break
    print("\nHeyyyyyy ",name,"\n ")
    print(''' \n    1) TOP DIRECTORS
    2)TOP_GENRES
    3)MOST_POPULAR_WORD
    4)HIGHEST_RETURN OF INVESTMENT
    5)AVERAGE RATING of top directors
    6)Top popular actors\n\n''')
    selection=int(input("\nEnter the option from above list: "))
   #makefile()
    if selection == 1 :
      Top_Director()
      director_name=input("\n\nEnter the director name for top 3 movies by rating: ")
      top_3_movies(director_name)
    elif selection == 2 :
      Top_Genres()
    elif selection == 3 :
      Most_popular_word()
    elif selection == 4 :
      Highest_ROI()
    elif selection==5 :
      Avg_rating()
    elif selection==6 :
      a=Popular_actor()
      k=int(input("\n\nEnter the number of top actors you require   "))
      print("\n\nFollowing are the list of top actors with number of movies acted\n ")
      for i, value in enumerate(a):
       if i < k:
        print(f"{i}: {value}")


if __name__ == "__main__":
  main()








```
